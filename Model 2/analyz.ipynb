```python
import pandas as pd
import matplotlib.pyplot as plt



df_spbu = pd.read_csv('spbu.csv')
df_msu = pd.read_csv('msu.csv')

df_spbu
```

```python
df_msu
```

```python
df_spbu['date'] = pd.to_datetime(df_spbu['date'])
df_spbu = df_spbu.groupby(pd.Grouper(key='date', freq='D')).agg({'likes':'sum', 'comments':'sum', 'reposts':'sum', 'views':'sum', 'date':'count'})
df_spbu = df_spbu.rename(columns={'date': 'posts'})
df_spbu.head(40)
```

```python
df_msu['date'] = pd.to_datetime(df_msu['date'])
df_msu = df_msu.groupby(pd.Grouper(key='date', freq='D')).agg({'likes':'sum', 'comments':'sum', 'reposts':'sum', 'views':'sum', 'date':'count'})
df_msu = df_msu.rename(columns={'date': 'posts'})
df_msu
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.plot(df_spbu.index, df_spbu['likes'], color='red', label='SPBU')
ax.plot(df_msu.index, df_msu['likes'], color='blue', label='MSU')

ax.set_title('Likes of VK per day')
ax.set_xlabel('Date')
ax.set_ylabel('Like')


ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.plot(df_spbu.index, df_spbu['comments'], color='red', label="SPBU")
ax.plot(df_msu.index, df_msu['comments'], color='blue', label="MSU")

ax.set_title('Comments of VK per day')
ax.set_xlabel('Date')
ax.set_ylabel('Comment')

ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(df_spbu.index, df_spbu['reposts'], color='red', label="SPBU")
ax.plot(df_msu.index, df_msu['reposts'], color='blue', label="MSU")

ax.set_title('Reposts of VK per day')
ax.set_xlabel('Date')
ax.set_ylabel('Repost')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(df_spbu.index, df_spbu['posts'], color='red', label='SPBU')
ax.plot(df_msu.index, df_msu['posts'], color='blue', label='MSU')
ax.set_title('Posts of VK per day')
ax.set_xlabel('Date')
ax.set_ylabel('Post')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(df_spbu.index, df_spbu['views'], color='red', label='SPBU')
ax.plot(df_msu.index, df_msu['views'], color='blue', label='MSU')
ax.set_title('Views of VK per day')
ax.set_xlabel('Date')
ax.set_ylabel('View')
ax.legend()
plt.show()
```
